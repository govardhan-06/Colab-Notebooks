### KDD Orange Cup Ensemble Selection for encoding categorical features

```python
for feature in df.columns:
    print(feature,":",len(df[feature].unique()))
```

```python
for feature in df.columns:
  c=10
  top_categories=df[feature].value_counts().sort_values(ascending=False).head(10).index
  for category in top_categories:
    df[category]=np.where(df[feature]==category,1,0)
  df.drop(feature,axis=1,inplace=True)
df.head()
```

### Probability Ratio Encoding

```python
df=pd.read_csv("train.csv")
df.head()
```

```python
df["Cabin"]=df["Cabin"].fillna("Missing")
```

```python
df["Cabin"]=[i[0] for i in df["Cabin"]]
```

```python
df["Cabin"].value_counts()
```

```python
prob=df.groupby(["Cabin"])["Survived"].mean()
prob
```

```python
prob_df=pd.DataFrame(prob)
```

```python
prob_df["Died"]=1-prob_df["Survived"]
```

```python
prob_df
```

```python
prob_df["Probability"]=prob_df["Survived"]/prob_df["Died"]
x=prob_df["Probability"].to_dict()
```

```python
x
```

```python
prob_df
```

```python
df['Cabin']=df['Cabin'].map(x)
```
